```python


# Get expected number of columns (from header)
with open(data_path, "r", encoding="utf-8") as file:
    header = file.readline().strip().split(",")
    expected_columns = len(header)

# Count bad lines
bad_lines_count = 0
with open(data_path, "r", encoding="utf-8") as file:
    for i, line in enumerate(file):
        if len(line.strip().split(",")) != expected_columns:
            print(f"‚ùå Bad line at row {i+1}: {line.strip()}")
            bad_lines_count += 1

print(f"\nüîπ Total bad lines found: {bad_lines_count}")
```
